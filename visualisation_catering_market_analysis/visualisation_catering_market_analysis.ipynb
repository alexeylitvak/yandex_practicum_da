## Предобработка данных
Переименуем столбец `number` и изменим значения в столбце `object_type`. Проверим данные на дубликаты. Проверим данные на корректность.

### Переименование
Переименуем столбец `number` и изменим значения в столбце `object_type`.

```python
# Переименуем столбец number
df.rename(columns={'number': 'seats'}, inplace=True)
```

```python
# Изменим значения в столбце object_type
df['object_type'].replace(
    {
        'предприятие быстрого обслуживания': 'фастфуд',
        'магазин (отдел кулинарии)': 'кулинария'
    }, inplace=True
)
```

### Обработка дубликатов
Проверим данные на явные строковые дубликаты.

```python
# Приведем столбец object_name к нижнему регистру
# для более точного поиска дубликатов
df['object_name'] = df['object_name'].str.lower()

# Проверим дубликаты
print('Дубликаты:', df.duplicated(
    subset=['object_name', 'chain', 'object_type', 'address', 'seats']
).sum())
```

Обнаружено 85 явных дубликатов, посмотрим на эти записи для того, чтобы понять дейсвительно ли эти записи дубликаты.

```python
# Выводим первые 30 записей-дубликатов
df[df.duplicated(
    subset=['object_name', 'chain', 'object_type', 'address', 'seats'], keep=False
)].sort_values(
    by=['address', 'object_name', 'seats']
).head(30)
```

При детальном анализе обнаружено:

* некоторые записи нельзя однозначно трактовать как дубликаты. К примеру, безымянное "кофе с собой" с нулем посадочных мест может быть как дублирующей записью, так и второй точкой в том же здании, относящейся к другому предпринимателю;
* кроме явных дубликатов есть еще неявные. К примеру, записи "пункт питания" и "пунк питания" или 2 заведения по одному адресу, но с указанием разных типов в `object_type`.

Ввиду того, что записей-дубликатов очень мало и не представляется возможным отделить настоящие дубликаты от случайных, удалять дубликаты не будем.

### Проверка данных на корректность
Проверим, верно ли указан тип заведения в столбце chain (сетевое или несетевое).

```python
# Выделяем заведения, где указан сетевой тип, но одно заведение
fake_chains = (
    df
    .query('chain == "да"')
    .groupby('object_name')
    .agg({'id': 'count'})
    .query('id == 1')
    .reset_index()
)
print(f'Некорректно тип для {len(fake_chains)} объектов')
```

```python
# Заменяем на несетевой  для найденных заведений
df.loc[df['object_name'].isin(fake_chains['object_name']), 'object_name'] = 'нет'
```

Исправлено указания типа заведения (сетевое или несетевое) для 331 объекта.

## Исследовательский анализ данных
Изучим сетевые и несетевые заведения, виды объектов общественного питания, число посадочных мест.

Для каждого заведения выделим улицу, определим топ-10 улиц по количеству заведений и улицы только с одним заведением. С помощью внешних данных по улицам найдем районы Москвы с большим и маленьким числом заведений общественного питания.

### Соотношение видов объектов общественного питания по количеству

```python
# Таблица видов объектов общественного питания по количеству
report = df.pivot_table(
    index='object_type',
    columns='chain',
    values='id',
    aggfunc='nunique'
)
report['Всего'] = report['да'] + report['нет']
report = (
    report
    .sort_values(by='Всего', ascending=False)
    .reset_index()
    .rename(columns={
        'object_type': 'Вид объекта общественного питания',
        'да': 'Сетевые',
        'нет': 'Несетевые',
    })
)
display(report[['Вид объекта общественного питания', 'Сетевые', 'Несетевые', 'Всего']])

# График видов объектов общественного питания по количеству
fig, ax = plt.subplots(figsize=(12, 6))

sns.set_color_codes('pastel')
sns.barplot(
    x='Вид объекта общественного питания',
    y='Несетевые',
    data=report,
    label='Несетевые',
    color='b'
)

sns.set_color_codes('muted')
sns.barplot(
    x='Вид объекта общественного питания',
    y='Сетевые',
    data=report,
    label='Сетевые',
    color='b'
)

ax.legend(loc='upper right', frameon=False)
ax.set(
    title='Виды объектов общественного питания по количеству',
    ylabel='Количество заведений'
)

sns.despine(left=True, bottom=True)
```

Как мы уже знаем, самым часто встречающимся в датасете видом объектов общественного питания является *кафе* (6099 заведений). На втором месте по распространенности — *столовые* (2587). Меньше всего в датасете представлены *кафетерии* (398), *закусочные* (360) и *кулинарии* (273).

### Cоотношение сетевых и несетевых заведений по количеству

```python
# Таблица сетевых и несетевых заведений по количеству
report = df.groupby('chain')['id'].count()
report = (
    report
    .reset_index()
    .rename(columns={
        'chain': 'Сетевой статус',
        'id': 'Количество заведений'
    })
         )
display(report)

# График видов объектов общественного питания по количеству
fig, ax = plt.subplots(figsize=(6, 6))

plt.pie(report['Количество заведений'], labels=['Сетевые', 'Несетевые'], colors=['#597dbf', '#abc9ea'], autopct='%.0f%%')

# Превращает pie-chart в donut-chart
donut = plt.Circle((0, 0), 0.70, facecolor='white')
fig = plt.gcf()
fig.gca().add_artist(donut)

ax.set(title='Соотношение сетевых и несетевых заведений')
plt.show()
```

В датасете только пятая часть заведений сетевые (2968 против 12398 несетевых заведений).

### Доля сетевых заведений по видам объектов общественного питания

```python
# Таблица долей сетевых заведений по видам объектов общественного питания 
report = df.pivot_table(
    index='object_type',
    columns='chain',
    values='id',
    aggfunc='nunique'
)
report['Доля сетевых заведений'] = report['да'] / (report['да'] + report['нет'])

report = (
    report
    .sort_values(by='Доля сетевых заведений', ascending=False)
    .reset_index()
    .rename(columns={
        'object_type': 'Вид объекта общественного питания',
    })
)
display(report[['Вид объекта общественного питания', 'Доля сетевых заведений']])

# График долей сетевых заведений по видам объектов общественного питания 
fig, ax = plt.subplots(figsize=(12, 6))

sns.set_color_codes('muted')
sns.barplot(
    x='Вид объекта общественного питания',
    y='Доля сетевых заведений',
    data=report,
    color='b'
)

ax.set(title='Доля сетевых заведений по видам объектов общественного питания')

sns.despine(left=True, bottom=True)
```

Наибольшая доля сетевых заведений в *фастфуде* (41%). Существенная доля сетей также среди *кулинарий* (29%), *ресторанов* (24%) и *кафе* (23%). Среди *баров*, *буфетов* и *столовых* сетей практически не встречается.

### Особенности числа посадочных мест для сетевых заведений

```python
# Собираем данные о сетях - название,
# число заведений, среднее число мест в заведении
chains = df.query('chain == "да"').groupby('object_name').agg({'id': 'count', 'seats': 'mean'})

# Определим границу (много/мало) заведений,
# посадочных мест с помощью медианы
chains['chain_size_group'] = 'Мало заведений'
chains.loc[chains['id'] > chains['id'].median(), 'chain_size_group'] = 'Много заведений'
chains['chain_seats_group'] = 'Мало мест'
chains.loc[chains['seats'] > chains['seats'].median(), 'chain_seats_group'] = 'Много мест'

# Добавим для каждого заведения одну из 4-х групп
chains['group'] = chains['chain_size_group'] + ' | ' + chains['chain_seats_group']

# Определяем и отфильтровываем выбросы
threshold = chains['id'].quantile(.99)
chains = chains.query('id < @threshold')

# График с разбивкой сетевых заведений по группам
# в зависимости от числа объектов и количетсва мест
fig, ax = plt.subplots(figsize=(12, 6))

sns.scatterplot(
    data=chains,
    x='id',
    y='seats',
    hue='group'
)

ax.legend(loc='upper right', frameon=False)
ax.set(
    title='Группы сетевых заведений по числу объектов и количеству посадочных мест',
    xlabel='Количество заведений',
    ylabel='Среднее число посадочных мест'
)

sns.despine()

# Таблица с разбивкой сетевых заведений по группам
# в зависимости от числа объектов и количетсва мест
(
    chains
    .reset_index()
    .groupby('group')
    .agg({
        'object_name': 'count',
        'id': 'median',
        'seats': 'median'
    })
    .reset_index()
    .rename(columns={
        'group': 'Группа',
        'object_name': 'Число сетей',
        'id': 'Медианное число заведений',
        'seats': 'Медианное число посадочных мест'
    })
    .style
    .format({'Медианное число посадочных мест': '{:.0f}'})
)
```

Для сетевых заведений характерно мало заведений с большим числом мест в каждом. Эта группа самая многочисленная (71 сеть). Медианное количество заведений в группе — 3, медианное число посадочных мест в группе — 93.

### Среднее количество посадочных мест

### Определение улицы

```python
# Определим функцию для выделения название улицы
def get_street_name(cell):
    '''
    Принимает адрес, разбивает на сегменты и возвращает
    тот сегмент адреса, где есть указание на улицу или
    второй сегмент, если такого указания не нашлось.
    '''
    for address_segment in cell.split(','):
        for street_type in [
            'бульвар',
            'набережная',
            'переулок',
            'площадь',
            'проезд',
            'проспект',
            'улица',
            'шоссе'
        ]:
            if street_type in address_segment:
                return address_segment.strip()
    return cell.split(',')[1].strip()
```

```python
# Выделим в отдельный столбец название улицы
df['street'] = df['address'].apply(get_street_name)

# Проверим, выведем 5 случайных строк датафрейма
df[['object_name', 'address', 'street']].sample(n=5, random_state=10)
```

Для каждого объекта улицу записали в столбец `street`.
