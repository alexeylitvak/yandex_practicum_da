### Распределение пользователей и платящих пользователей по странам

```python
# Функция для группировки датафрейма по всем
# пользователям или только по платящим пользователям
def group_profiles(df, dimension, payer=False):
    '''
    Получает датафрейм с пользовательскими профилями
    и флаг делать ли срез платящих пользователей.
    Возвращает сгруппированный датафрейм по всем
    пользователям или только по платящим пользователям.
    '''
    if payer:
        df = df.query('payer == True')
    df_grouped = (
        df
        .groupby(dimension)['user_id']
        .count()
        .sort_values(ascending=True)
    )
    return df_grouped
```

```python
# Функция для построения диаграмм распределения
# пользователей и платящих пользователей
def get_profiles_bars(df, dimension, dimension_title):
    '''
    Получает датафрейм с пользовательскими профилями,
    параметр для группировки, название параметра для подписи. 
    Возвращает диаграммы распределения пользователей
    и платящих пользователей по указанному параметру.
    '''

    fig, axs = plt.subplots(1, 2, figsize=(13, 5))

    df_grouped = group_profiles(df, dimension)
    
    df_grouped.plot(ax=axs[0], kind='barh')
    axs[0].set_title(f'Пользователи {dimension_title}')
    axs[0].set(xlabel='пользователи',
               ylabel='страна')
    axs[0].margins(x=0.15)
    
    # Подписываем значения
    for i, val in enumerate(df_grouped):
        axs[0].text(
            val+500, i, val,
            verticalalignment='center',
            color='gray')

    df_grouped = group_profiles(df, dimension, payer=True)
    
    df_grouped.plot(ax=axs[1], kind='barh')
    axs[1].set_title(f'Платящие пользователи {dimension_title}')
    axs[1].set(xlabel='платящие пользователи',
               ylabel='страна')
    axs[1].margins(x=0.15)
    
    # Подписываем значения
    for i, val in enumerate(df_grouped):
        axs[1].text(
            val+50, i, val,
            verticalalignment='center',
            color='gray')

    plt.tight_layout()
    plt.show()
```

```python
# Функция для составления таблицы с процентом платящих пользователей
def get_profiles_table(df, dimension):
    '''
    Получает датафрейм с пользовательскими профилями,
    параметр для группировки. Возвращает таблицу
    с группировкой по указанному параметру, количеством
    пользователей и процентом платящих пользователей.
    Выделяет максимальные значения столбцов.
    '''
    profiles_table = (
        df
        .groupby(dimension)['user_id', 'payer']
        .agg({'user_id': 'count', 'payer': 'mean'})
        .sort_values(by='payer', ascending=False)
        .style
        .format({'payer': '{:.2%}'})
        .highlight_max(axis=0, color='lightyellow')
    )
    return profiles_table
```

```python
# Распределение расходов по рекламным каналам
cost_grouped = (
    costs
    .groupby('channel')['costs']
    .sum()
    .sort_values()
)
    
cost_grouped.plot(kind='barh', figsize=(8, 6))

# Подписываем значения
for i, val in enumerate(cost_grouped):
        plt.text(
            val+500, i, round(val),
            verticalalignment='center',
            color='gray')

plt.title('Распределение расходов по рекламным каналам')
plt.xlabel('расходы')
plt.ylabel('рекламный канал')
plt.margins(x=0.1)
plt.tight_layout()
plt.show()
```

Самые крупные расходы маркетинга — привлечение трафика из каналов TipTop и FaceBoom.

В динамике рассмотрим крупные рекламные каналы (TipTop и FaceBoom) и объединенные данные по остальным 8 рекламным каналам (other). Для удобства чтения графика применим сглаживание с недельным окном с помощью скользящего среднего.

```python
# Объединяем прочие рекламные каналы в other
costs_dymanic = costs.copy()
costs_dymanic.loc[(costs['channel'] != 'TipTop')
                  & (costs['channel'] != 'FaceBoom'),
                  ['channel']] = 'other'

# Строим сводную таблицу с расходами по 3 каналам в динамике
costs_dymanic = (
    costs_dymanic
    .pivot_table(
        index='dt',
        columns='channel',
        values='costs',
        aggfunc='sum'
    )
)

# Ширина окна сглаживания
window = 7

# Для каждого канала применяем скользящее среднее
for column in costs_dymanic.columns.values:
    costs_dymanic[column] = (
        costs_dymanic[column]
        .rolling(window)
        .mean()
    )

# Динамика расходов по рекламным каналам
(
    costs_dymanic.iloc[6:] # отрезаем образовавшиеся пустые строки
    .plot(grid=True, figsize=(14, 6))
)
plt.title('Динамика расходов по рекламным каналам')
plt.xlabel('')
plt.ylabel('расходы')
plt.legend(title='Рекламный канал')
plt.show()
```

С июня расходы на рекламные каналы Faceboom и TipTop становятся самыми крупными маркетинговыми расходами. Вероятно, отдел маркетинга решил, что эти каналы наиболее эффективны и перераспределил рекламный бюджет в пользу Faceboom и TipTop.

Расходы на все рекламные кампании возрастают с середины каждого месяца и падают с началом нового, кроме периода август-сентябрь (держатся на постоянном уровне). Такая тенденция может быть объяснена оптимизацией рекламных бюджетов дважды в месяц.

В целом видно, что расходы по рекламному каналу TipTop продолжают повышаться, а то время как расходы по TipTop и прочим каналам остаются на постоянном уровне (не считая пиков начала месяца).
